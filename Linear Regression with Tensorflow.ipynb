# Linear Regression with Tensorflow

## : to understand easily

1) Import tensorflow as tf and check out the version

```python
import tensorflow as tf
tf.__version__
```

2) Make x, y data set and random variable W(weight), b(bias)

```python
import tensorflow as tf

#X and y data
x_train = [i for i in range(10)]
y_train = [i for i in range(10)]

W=tf.Variable(tf.random_normal([1]), name='weight') # Make random variable in 1-dimentional array
b=tf.Variable(tf.random_normal([1]), name='bias')

#hyphothesis
hypo=x_train * W +b
```

3) Make cost function

```python
#cost function
cost=tf.reduce_mean(tf.square(hypo-y_train))
```

4) Set Optimizer, learning_rate

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

5) Set sess and Do run

```python
sess = tf.Session()

sess.run(tf.global_variables_initializer())

#check the result every 20 times
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost),sess.run(W),sess.run(b))
        
# When W is close to 1 and b is close to 0, it goes right way!!!
```
